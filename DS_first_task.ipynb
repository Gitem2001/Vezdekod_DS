# Создание датафрейма и его заполнение

```python
df = pd.DataFrame(columns=['text','likes','comments','views','reposts'])
```

```python
for i in dict(data.json())['response']['items']:   # i - определенный item  в данном случае пост
  try:  
    row = {'text':i['text'],'likes':i['likes']['count'],'comments':i['comments']['count'],'views':i['views']['count'],'reposts':i['reposts']['count']}    # новая строка для добавления в датафрейм
  except:
    row = {'text':i['text'],'likes':i['likes']['count'],'comments':i['comments']['count'],'views':0} 

  df = df.append(row,ignore_index=True)
```

```python
for i in df.columns: # проверка на количество nan
  print(i,df[i].isna().sum()) 
```
